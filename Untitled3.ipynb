```python
import numpy as np # To work with arrays
import pandas as pd #to work with the data
from sklearn.model_selection import train_test_split #module to split the data into training and testing
from sklearn.linear_model import LogisticRegression #to build model
from sklearn.metrics import accuracy_score #Evaluation
```

```python
#importing the data
df=pd.read_csv('sample_data/heart_disease_data.csv')
```

**Splitting Features and Target**

```python
x=df.drop(columns="target",axis=1)
y=df["target"]
```

**Training and Testing**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

```python
y.shape,y_train.shape,y_test.shape
```

**Model Building**

LogisticRegression is used to create the model where end of the it is a binary classification

```python
model=LogisticRegression()
```

```python
#train the model

model.fit(x_train,y_train)
```

**Evaluation**

```python
# accuracy_score on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print("Accuracy on training data : ",training_data_accuracy)
```

```python
# accuracy_score on testing data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
print("Accuracy on testing data : ",testing_data_accuracy)
```

**Building the Predicting System**

```python
input_data=(49,1,1,130,266,0,1,171,0,0.6,2,0,2)# Tuple of the feature data-->which is my input
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
if(prediction[0]==0):
  print("The person does not have a heart disease")
else:
  print("The person has a heart disease")
```
